```python
from finalytics import Portfolio
```

```python
ticker_symbols = ["AAPL", "MSFT", "GOOG", "BTC-USD"]
benchmark_symbol = "^GSPC"
start_date = "2019-01-01"
end_date = "2023-11-01"
interval = "1d"
confidence_level = 0.95
risk_free_rate = 0.04
max_iterations = 1000
objective = "max_sharpe"
```

```python
portfolio = Portfolio(ticker_symbols, 
                      benchmark_symbol, 
                      start_date, 
                      end_date, 
                      interval, 
                      confidence_level, 
                      risk_free_rate, 
                      max_iterations, 
                      objective)
```

```python
portfolio.get_optimization_results()
```

```python
portfolio.display_portfolio_charts("png")
```
